# Does Bitcoin behave as a currency?

## First approach: Bitcoin evolution vs 3 main currencies

I checked the evolution of the Bitcoin index as well as the evolution for other 3 main indexes in the world: US Dollar, Euro and Japanese Yen. 

At their own scale, the three currencies are quite stable, while the Bitcoin evolution is not, with two main peaks in December 2017 and July in 2019 followed by pronounced falls.

There's no clear realtionship between the evolution of Bitcoin and the evolution of these three main currencies.

## Second approach: Bitcoin evolution vs 7 tech stocks

After not finding a clear relationship between Bitcoin and currencies trends, I decided to compare the Bitcoin evolution to the evolution of some tech companies stocks. For that analysis, I focused on worldwide tech companies such as Amazon, Google, Samsung and 4 more

They have different scales, but there are clearly two groups. On one side Samsung, and Amazon. On the other side, Google, Facebook, Alibaba, Apple and Microsoft. I focused on the first group, as they have some smooth relationship to the evolution of Bitcoin at some points. But it seems that the tech stocks follows the pattern of Bitcoin and not viceversa as I expected. However, this is not a strong relation between them.

## The Bitcoin peak

At this point, I wondered what could happen in December 2017 and January 2018 that caused a fast increasing of Bitcoin value followed by a fast fall of it. One of the news I found was that the China Government forbidded the ICO's (Initial Coin Offering) and China was one of the then-dominant bitcoin exchanges country.

A new question appeared: could China's currency be related with Bitcoin?

## Third approach: Bitcoin Evolution vs Chinese Yuan

As I couldn't find the evolution of Chinese Yuan index for all the timelapse studied, I focused on the Exchange Rate between Chinese Yuan and US Dollar.

And surprisingly, the plot of these two trends was the most clearly related I got. But the relationship changes with the time. Since de beggining of the study (January 2017) the relationship was negative, that means that when the Yuan-Dollar exchange rate was falling, Bitcoin index was raising, and viceversa.

But about November 2018 this trend changed and from then on, the relationship seems to be turned into positive, behaving both in the same way.

## Conclusion

It is clear that Bitcoin does not behave as a currency, but it is also clear that some government decisions that may affect currencies, may also affect Bitcoin evolution, as seen with Chinese Government decision on ICO's.

But there is lots of news, studies and guesses about what happened, so it is not possible to make an unequivocal statement about it.

It is necessary more investigation to know which factors, or decisions, could have bigger effects on Bitcoin evolution.
